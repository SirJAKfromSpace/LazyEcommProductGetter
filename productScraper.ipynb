# Web Scrape Ecommerce Site Products

### Scrape list of product category links

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = 'http://Ecommerce.com/'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
listofA = soup.findAll('a', {'class': 'site-map-block__list'})
categorysitemap = []
for anchorelem in listofA:
    categorysitemap.append(anchorelem['href'])
    
print('Number of Cat Pages:',len(categorysitemap))
categorysitemap
```

```python
listofproducts = []
catcount = 0
start_time = time.time()
print('Scraping Started at', time.ctime())
for site in categorysitemap:
    res = requests.get(site)
    if(res.status_code==200):
        soup = BeautifulSoup(res.text, 'html.parser')
        catcount += 1
        breadcrumbs = soup.findAll('ul',{'class': 'items'})[0].findChildren() # get category and subcat from breadcrumb
        catname = breadcrumbs[2].text.strip(' ')
        subcatname = breadcrumbs[4].text.strip(' ')
        print(catname+' : '+subcatname, '['+str(catcount)+'/'+str(len(categorysitemap))+']')
        pagenum = 0
        while True:
            productancs = soup.findAll('a', {'class': 'product-item-link'})
            for anchorelem in productancs:
                listofproducts.append([anchorelem['href'], catname, subcatname])
            try:
                pagenum += 1
                print(pagenum,end='-')
                nextpage = soup.findAll('a', {'class': 'next'})[0]['href']
                res = requests.get(nextpage)
                soup = BeautifulSoup(res.text, 'html.parser')
                if (res.status_code!=200):
                    print('REQ ERROR', res.status_code)
                    break
            except:
                print('END')
                print('Scraped',len(listofproducts),'Product links')
                df = pd.DataFrame(listofproducts, columns =['link', 'catname', 'subcatname'])
                df.to_csv('ecommerce_productlist.csv', index=False)
                print('CSV Saved for',site.split('/')[-1].strip('.html'))
                print('')
                break
print('Scraping Complete at', time.ctime())
print('Time Elapsed: ', time.strftime("%H hrs %M mins %S secs", time.gmtime(time.time() - start_time)))
```

```python
df = pd.read_csv('ecommerce_productlist.csv')
df
```

### Scrape Individual Products from Product Links List

```python
itemnamearr = []
itempricearr = []
itemqtyarr = []
itemcascadearr = []
totalprod = len(df)
start_time = time.time()
print('Scraping Started at', time.ctime())
for index, row in df.iterrows():
    res = requests.get(row['link'])
    if(res.status_code==200):
        soup = BeautifulSoup(res.text, 'html.parser')
        itemname = soup.findAll('h1',{'class': 'page-title'})[0].findChild().text
        try:
            itemprice = soup.findAll('span',{'class': 'price'})[0].text
        except IndexError:
            itemprice = 'null'
        try:
            itemqty = soup.findAll('input',{'class': 'input-text qty'})[0]['value'] if len(soup.findAll('input',{'class': 'input-text qty'}))>=1 else soup.findAll('span',{'class': 'tier-select-qty'})[0].text
        except IndexError:
            itemqty = 'null'
        try:
            t2 = ''
            for node in soup.findAll('tr',{'class': 'item'}):
                t = ('_'.join(node.findAll(text=True)))
                for i,c in enumerate(t):
                    if(c=='%'):
                        t = t[:i-1]+t[i:]
                t = t.replace(' ','').replace('_',' ')[1:].replace(' - ','-').replace('  ','').replace(' +','+')
                t2 += t+' | '
            itemcascade = t2
        except IndexError:
            itemcascade = 'null'
            
        itemnamearr.append(itemname)
        itempricearr.append(itemprice)
        itemqtyarr.append(itemqty)
        itemcascadearr.append(itemcascade)
    else:
        itemnamearr.append('null E'+str(res.status_code))
        itempricearr.append('null')
        itemqtyarr.append('null')
        itemcascadearr.append('null')
        print('NullError, Code:',res.status_code,'Link:',row['link'])
    print('[{}/{}] Product Pages Scraped'.format(index+1,totalprod), end='\r')
    
dftemp = df.assign(name=itemnamearr, price=itempricearr, quantity=itemqtyarr, qpcascade=itemcascadearr)
dftemp = dftemp[['name','price','quantity','qpcascade''catname','subcatname','link']]
print('Scraping Complete at', time.ctime())
print('Time Elapsed: ', time.strftime("%H hrs %M mins %S secs", time.gmtime(time.time() - start_time)))
dftemp
```

```python
dftemp.to_csv('ecommerce_products.csv',index=False)
```
